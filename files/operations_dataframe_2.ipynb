```python
import pandas as pd
import numpy as np
from faker import Faker
import random
from datetime import datetime, timedelta
import string

# Initialize Faker
fake = Faker()

# Set the date range
start_date = datetime(2020, 1, 1)
end_date = datetime(2023, 12, 31)

# Generate driver data
num_drivers = 700
driver_info = {
    'Driver_id': list(range(1, num_drivers + 1)),
    'Driver_name': [fake.name() for _ in range(num_drivers)],
    'Marital_status': np.random.choice(['Single', 'Married', 'Divorced', 'Widowed'], size=num_drivers),
    'Age': np.random.randint(25, 65, size=num_drivers),
    'Education': np.random.choice(['Elementary School', 'Middle School', 'High School', 'Bachelor'], size=num_drivers)
}

# Create a dataframe for drivers
drivers_df = pd.DataFrame(driver_info)


# Generate customer data
num_customers = 1850
customer_info = {
    'Customer_id': list(range(1, num_customers + 1)),
    'Customer_name': [fake.company() for _ in range(num_customers)],    
}

# Create a DataFrame for customers
customers_df = pd.DataFrame(customer_info)


# Generate trailer data

# Function to generate trailer code
def generate_trailer_code(i):
    return f"TR-{i:06d}"

# Function to generate random year between 2000 and 2024
def generate_year():
    return random.randint(2000, 2024)

# Sample data for makes and axle configurations
makes = ['John Deere', 'Caterpillar', 'Komatsu', 'Volvo', 'Hitachi']
axle_configurations = ['4x2', '6x4', '8x6']

# Generate trailer data
num_trailers = 900
trailer_info = {
    'Trailer_id': list(range(1, num_trailers + 1)),
    'Trailer_code': [generate_trailer_code(i) for i in range(1, num_trailers + 1)],
    'Trailer_year_model': [generate_year() for _ in range(num_trailers)],
    'Make': [random.choice(makes) for _ in range(num_trailers)],
    'Horsepower': [random.randint(200, 600) for _ in range(num_trailers)],
    'Axle_configuration': [random.choice(axle_configurations) for _ in range(num_trailers)]
}

# Create DataFrame
trailers_df = pd.DataFrame(trailer_info)


# Generate trip data
num_trips = 300000
departure_dates = [fake.date_between_dates(start_date, end_date) for _ in range(num_trips)]
trip_info = {
    'Trip_id': list(range(1, num_trips + 1)),
    'Departure_date': departure_dates,
    'Driver_id': [np.random.randint(1, num_drivers + 1) for _ in range(num_trips)],
    'Trailer_id': [np.random.randint(1, num_trailers) for _ in range(num_trips)],
    'Customer_id': [np.random.randint(1, num_customers) for _ in range(num_trips)],
    'Origin_location': [fake.city() + ', ' + fake.state_abbr() for _ in range(num_trips)],
    'Destination_location': [fake.city() + ', ' + fake.state_abbr()for _ in range(num_trips)],
    'Km_travelled': np.random.randint(50, 5000, size=num_trips),
    'Diesel_expenses': np.random.uniform(50, 200, size=num_trips),
    'Trailer_status': np.random.choice(['Active', 'Inactive', 'Under Maintenance'], size=num_trips),
    'Service_category': np.random.choice(['Import', 'Export', 'Local', 'Gasoline', 'Coal', 'Gravel', 'Non-metallic mineral products', 'Cereal', 'Special Customers', 'Other'], size=num_trips),
    'Road_status': np.random.choice(['Normal', 'Blocked', 'Under Construction', 'Closed', 'Impassable'], size=num_trips),
    'Trip_status': np.random.choice(['Scheduled', 'To Leave', 'In Transit', 'Arrived', 'Ended', 'Empty Return'], size=num_trips),
    'Invoice_status': np.random.choice(['Paid', 'Pending'], size=num_trips),
    'Invoice_number': random.sample(range(10000000, 99999999), num_trips),
    'Income_per_km': np.random.uniform(19, 35, size=num_trips)
}

# Initialize an empty list for Arrival_date
trip_info['Arrival_date'] = []

# Calculate days travelled based on Km travelled
for trip_id in range(1, num_trips + 1):
    km_travelled = trip_info['Km_travelled'][trip_id - 1]
    days_travelled = int(km_travelled // (1046 + 1))
    # trip_info['days_travelled'].append(days_travelled)

    # calculate arrival date
    departure_date = trip_info['Departure_date'][trip_id - 1]
    arrival_date = departure_date + timedelta(days=days_travelled)
    trip_info['Arrival_date'].append(arrival_date)

# Create a dataframe for trips
trips_df = pd.DataFrame(trip_info)
```
